```python
import matplotlib.pyplot as plt
```

```python
max_pop = 60000000
infected = 500000
t=[]
cases=[]
total=[]
infecous=[0,0,0,0,100000,100000,100000,100000,100000]
for i in range(150):
    t.append(i)
    new = int(sum(infecous)*0.15*(1-(infected/max_pop)))
    infecous.pop(0)
    infecous.append(new)
    infected+=new
    cases.append(new)
    total.append(infected)
    print(i, infected,new)
```

```python
plt.plot(cases)
plt.show()
plt.plot(total)
plt.show()
max_case = max(cases)
pos = cases.index(max_case)
print(pos, max_case, int(max_case*0.02))
```
